```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as matplotlib
import requests
from bs4 import BeautifulSoup as bs
import lxml
```

```python
#reading data extracted from the url into dataframe
hecjournals=pd.read_csv("hecjournals.csv")
hecjournals_copy=hecjournals.copy()
hecjournals.head()
```

```python
#reading the dataset that was already given to us.. 
eprint=pd.read_csv("eprint_202112111250.csv")
eprint_copy=eprint.copy()
eprint
```

# Cleaning Dataframes

```python
print('Number of null values in eprint dataframe before cleaning: ', eprint.isnull().sum().sum())
print('Number of null values in hecjournals dataframe before cleaning: ', hecjournals.isnull().sum().sum())
```

```python
eprint_clean=pd.DataFrame(eprint.dropna(how='all',thresh=600, axis=1)) #drop all columns that contain more than 600 null entries
hecjournals_clean=pd.DataFrame(hecjournals.dropna(how='all', thresh=3000, axis=1)) #drop all columns that contain more than 3000 null entries
print('Number of null values in eprint dataframe after cleaning: ', eprint_clean.isnull().sum().sum())
print('Number of null values in hecjournals dataframe after cleaning: ', hecjournals_clean.isnull().sum().sum())
```

```python
eprint_clean.to_csv("eprint_202112111250_clean.csv")
hecjournals_clean.to_csv("hecjournals_clean.csv")
```

# Mapping

```python
#s = eprint_copy.eprintid.replace(hecjournals['Journal name'])
hecjournals_copy.rename({'Journal name':'Title'}, axis=1, inplace=True)
s = hecjournals_copy.Title.replace(eprint_copy.set_index('eprintid')['title'])
s
```

```python
eprint_copy.rename({'eprintid': 'Journal_name'},axis=1, inplace=True)
eprint_copy['Journal_name']=s
eprint_copy[['Journal_name', 'title']]
```

```python
hecjournals_copy['eprint_title']=s
hecjournals_copy[['eprint_title', 'title']]
```

# Merging

```python
anyk=pd.DataFrame()
anyk['titles_all']=hecjournals_copy['eprint_title'].append(hecjournals_copy['title'].dropna())
anyk.index = pd.RangeIndex(len(anyk.index))
anyk
```
